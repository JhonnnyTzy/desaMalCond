```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definimos el rango de valores para x y y
x = np.linspace(0, 2, 100)
y = np.linspace(0, 2, 100)

# Creamos una malla de puntos (x, y)
X, Y = np.meshgrid(x, y)

# Definimos las ecuaciones del sistema en funcin de z
Z1 = (3 - 1.0001 * X - 1.0000 * Y) / 0.9999
Z2 = (3 - 1.0000 * X - 1.0000 * Y) / 1.0000
Z3 = (3 - 0.9999 * X - 1.0000 * Y) / 1.0001

# Creamos la grfica 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Graficamos las superficies
ax.plot_surface(X, Y, Z1, cmap="Blues", alpha=0.6)
ax.plot_surface(X, Y, Z2, cmap="Reds", alpha=0.6)
ax.plot_surface(X, Y, Z3, cmap="Greens", alpha=0.6)

# Aadimos etiquetas y ttulo
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Sistema de ecuaciones lineales 3x3 con pequeos cambios en los coeficientes')

# Mostramos la grfica
plt.show()

```
